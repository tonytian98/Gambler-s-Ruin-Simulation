```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# probability of winning and losing
p=0.4
q=1-p

#initial money
money=10

#winning losing amount
w=1
l=-1

#maximum steps allowed
maxi=50

#Number of Iteration
numit=10
```

```python
tracks=[]

for i in range(numit):
    m=money
    track=[]
    track.append(m)
    while m!=0 and len(track)!=maxi+1:
        o=np.random.choice([w,l],p=[p,q])
        m=m+o
        track.append(m)
    tracks.append(track)
    
```

```python
plt.figure(figsize=(int(maxi/3),int(money/3)))
plt.xticks(np.linspace(0,maxi,maxi+1))
for track in tracks:

    plt.plot(np.linspace(0,len(track)-1,len(track)),track)

    plt.xlabel('Number Of Games')
    plt.ylabel('Amount Of Money')
    plt.title("Gambler's ruin (P(winning)={}) \n #Iteration={} ".format(p,numit))
```
