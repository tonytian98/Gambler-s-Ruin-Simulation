```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# probability of winning and losing
p=0.4
q=1-p

#initial money
m=10

#winning losing amount
w=1
l=-1
```

```python
track=[]
track.append(m)
while m!=0 and len(track)!=50:
    o=np.random.choice([w,l],p=[p,q])
    m=m+o
    track.append(m)
    
```

```python
plt.figure(figsize=(int(len(track)/3),int(max(track)/3)))
plt.plot(np.linspace(0,len(track),len(track)),track)
plt.xticks(np.linspace(0,len(track),len(track)+1))
plt.xlabel('Number Of Games')
plt.ylabel('Amount Of Money')
plt.title('P(winning)={} '.format(p))
```
